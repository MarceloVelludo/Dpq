```python
import DPQ
import numpy as np
import QuantumTomography as qKLib
```

```python
dpq = DPQ.DinamicaPontosQuanticos()
```

```python
#definindo os parametros para a hamiltoniana
mhz = 10**(-3)
pi = np.pi
j_1 = 280*mhz*2*pi
j_2 = 320*mhz*2*pi
bz_1 = (pi/16)*(10**3)*mhz
bz_2 = (pi/16)*(10**3)*mhz
j12 = pi/140
h = dpq.hamiltoniana_Nova(j_1, j_2, bz_1, bz_2, j12)
#Esse Ã‰ O CENTRAL, variar o bz1 e o bz2 com 20%
#Mantendo o j1 e j2 fixos
```

```python
h
```

```python
ro = dpq.ro(h, 800)
#dpq.concurrence(ro)

print("Minha Concurrence:", dpq.concurrence(ro))
print("Concurrence:", qKLib.concurrence(ro))
print("Entropy:", qKLib.entropy(ro))
```

```python
tempos = np.arange(1500, step=1)
```

```python
concurrence = []
for tempo in tempos:
    ro = dpq.ro(h, tempo)
    concurrence.append(dpq.concurrence(ro))
```

```python
import matplotlib.pyplot as plt

x = concurrence
y = tempos

fig, ax = plt.subplots(figsize=(20,10))

# Using set_dashes() to modify dashing of an existing line
line1, = ax.plot(y,x , label=' ')

ax.legend()
plt.show()
```
